```python
import pandas as pd

# Baca dataset (ganti dengan nama file yang sesuai)
df = pd.read_csv("/content/data/austin_weather.csv")

#Deksripsi
print(df.describe())

print(df.info())
```

# Soal 3

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd



#Max
plt.figure(figsize=(10, 6))
sns.histplot(df['TempHighF'], kde=True)
plt.title('Distribusi Suhu Maksimum di Basel (BASEL_temp_max)')
plt.xlabel('Suhu Maksimum (째C)')
plt.ylabel('Frekuensi')
plt.show()


#Min
plt.figure(figsize=(10, 6))
sns.histplot(df['TempLowF'], kde=True)
plt.title('Distribusi Suhu Minimum')
plt.xlabel('Suhu Minimum(째C)')
plt.ylabel('Frekuensi')
plt.show()

#Perbandingan
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TempHighF', y='TempLowF', data=df)
plt.title('Hubungan antara Suhu Maksimum dan Minimum')
plt.xlabel('Suhu Maksimum (째C)')
plt.ylabel('Suhu Minimum (째C)')
plt.show()
```

# Soal 4

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd



# Scatter plot dengan hue 'Events'
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x=df['TempHighF'],
    y=df['TempLowF'],
    hue=df['Events'],           # Menggunakan kolom 'Events' untuk hue
    palette='Set1',             # Palet warna untuk membedakan kondisi cuaca
    alpha=0.7                   # Transparansi titik
)
plt.title('Scatter Plot Suhu Maksimum vs Minimum di Austin berdasarkan Kondisi Cuaca')
plt.xlabel('Suhu Maksimum (째F)')
plt.ylabel('Suhu Minimum (째F)')
plt.legend(title='Kondisi Cuaca', loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

# Soal 6

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



plt.figure(figsize=(10, 6))
# Menggunakan palet warna 'husl' dan rotasi label 30 derajat
sns.countplot(data=df, x='Events', order=df['Events'].value_counts().index, palette='viridis')

plt.title('Frekuensi Kondisi Cuaca di Austin', fontsize=16)
plt.xlabel('Kondisi Cuaca', fontsize=12)
plt.ylabel('Frekuensi', fontsize=12)
plt.xticks(rotation=30, ha='right')

plt.show()



```

```python
import pandas as pd
import matplotlib.pyplot as plt

print(df.dtypes)

plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['TempHighF'], color ='orange', label='Suhu Maksimum (째F)')

plt.title('Perubahan Suhu Maksimum dengan Garis Tren di Austin', fontsize=16)
plt.xlabel('Tanggal', fontsize=12)
plt.ylabel('Suhu Maksimum (째F)', fontsize=12)
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


plt.figure(figsize=(12, 6))

# Membuat boxplot
sns.boxplot(x='Events', y='TempHighF', data=df, palette='Set2')

plt.title('Boxplot Suhu Maksimum Berdasarkan Kondisi Cuaca di Austin', fontsize=16)
plt.xlabel('Kondisi Cuaca', fontsize=12)
plt.ylabel('Suhu Maksimum (째F)', fontsize=12)
plt.xticks(rotation=30, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```
